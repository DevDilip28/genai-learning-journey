# 📘 Embeddings 

## 🔹 Definition
Embeddings convert token IDs into dense vectors that capture meaning.

👉 Token → Vector (meaning)

---

## 🔹 Why Needed?
Token IDs are just numbers (no meaning ❌)

Example:
"cat" → 123  
"dog" → 456  

👉 No relationship

✔ Embeddings solve this:
"cat" → [0.2, 0.8, ...]  
"dog" → [0.25, 0.75, ...]

👉 Similar words → similar vectors

---

## 🔹 Intuition (Most Important)

Think of words placed in space:

- king near queen  
- dog near cat  
- apple far from car  

👉 Each word = coordinates (vector)

---

## 🔹 What is a Vector?

A list of decimal numbers:

[0.12, -0.45, 0.88, ...]

✔ High-dimensional (e.g., 768)  
✔ Represents meaning numerically  

---

## 🔹 Embedding Matrix

- Shape: (V × D)

Where:
- V = vocabulary size  
- D = embedding dimension  

👉 Each row = vector of one token  

---

## 🔹 How it Works

Token ID = 123  
↓

Pick row 123 from embedding matrix  
↓

That row = embedding vector  

👉 Simple lookup (no heavy computation)

---

## 🔹 Example

Token IDs:
[45, 12, 98]

↓

Vectors:
dog → [0.21, 0.77, ...]  
is → [0.11, 0.05, ...]  
running → [0.67, 0.32, ...]

👉 Model works on vectors, not words

---

## 🔹 Key Properties

- Similar words → similar vectors  
- Captures semantic relationships  
- Example:
  king - man + woman ≈ queen  

---

## 🔹 Types of Embeddings

1. Token Embedding → meaning  
2. Positional Embedding → position  

👉 Both combined before transformer

---

## 🔹 Where Used

### 1. LLMs
Tokens → embeddings → transformer  

### 2. RAG
Text → embeddings → vector DB → similarity search  

### 3. Embedding APIs
Used for:
- Search  
- Clustering  
- Recommendation  

---

## 🔹 Pipeline So Far

Text  
↓  
Tokens  
↓  
Token IDs  
↓  
🔥 Embeddings  
↓  
Transformer  

---

## 🔹 Key Insight
Token IDs = meaningless  
Embeddings = meaning

---

## 🔹 One-Line Summary
Embeddings convert token IDs into meaningful high-dimensional vectors that models use for understanding.

```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModel.from_pretrained("gpt2")

text = "dog is running"

inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)

embeddings = outputs.last_hidden_state

print(inputs)
print(outputs)
print(embeddings) 
```

```python
from transformers import AutoTokenizer, AutoModel
import torch 

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModel.from_pretrained("gpt2")

text = "Dilip"

input = tokenizer(text, return_tensors="pt")
output = model(**input)

embedding = output.last_hidden_state

print(embedding)
```
