# 📘 Tokenization 

## 🔹 Definition
Tokenization = converting text into smaller units (tokens) for model processing.

---

## 🔹 Why Needed?
Computers don’t understand text, only numbers.

Text → Tokens → Token IDs → Model

---

## 🔹 What are Tokens?
Tokens can be:
- Words
- Subwords ✅ (used in LLMs)
- Characters

---

## 🔹 Types of Tokenization

### 1. Word-level
"I love AI" → ["I", "love", "AI"]  
❌ Large vocab, unknown words problem

### 2. Character-level
"cat" → ["c", "a", "t"]  
❌ Too long, inefficient

### 3. Subword (LLMs use this ✅)
"unbelievable" → ["un", "believ", "able"]  
✔ Best balance (efficiency + flexibility)

---

## 🔹 Subword Methods
- BPE (Byte Pair Encoding)
- WordPiece
- SentencePiece

👉 Idea: merge frequent character pairs to form meaningful subwords

---

## 🔹 Example

"ChatGPT is powerful"

→ Tokens:
["Chat", "G", "PT", " is", " powerful"]

→ Token IDs:
[1234, 567, 890, 45, 6789]

---

## 🔹 Important Points
- Tokens ≠ always words
- Spaces can be part of tokens
- Same word can split differently
- Tokenization depends on model

---

## 🔹 LLM Pipeline

Text  
↓  
Tokens  
↓  
Token IDs  
↓  
Embeddings  
↓  
Transformer  

---

## 🔹 Tokens & Cost
- More tokens = more cost 💸
- Context window = max tokens model can handle

---

## 🔹 Key Insight
LLMs do NOT see words  
They only see token IDs

---

## 🔹 One-Line Summary
Tokenization converts text into numerical units (tokens & IDs) so models can process it.

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")

text = "ChatGPT is Powerfull"

tokens = tokenizer.tokenize(text)
ids = tokenizer.encode(text)

print(tokens)
print(ids)
```
