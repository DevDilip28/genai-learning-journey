## 🌐 Backend Foundations for GenAI

### 1. HTTP Basics

### 🧠 Intuition
HTTP is the **communication protocol of the web**.

- Client → sends **request**
- Server → sends **response**

👉 Think:
Client (you) → Request (order food)  
Server       → Response (deliver food)


### ⚙️ Core Concepts

#### Request
- **Method** → action (GET, POST, etc.)
- **URL** → where to send request
- **Headers** → metadata (auth, content type)
- **Body** → actual data (optional)

#### Response
- **Status Code**
- **Data** (JSON, HTML, etc.)

### 🔁 HTTP Methods

| Method | Purpose        |
|--------|---------------|
| GET    | Fetch data    |
| POST   | Send data     |
| PUT    | Update data   |
| DELETE | Remove data   |


### 📊 Status Codes

| Code | Meaning        |
|------|---------------|
| 200  | OK            |
| 201  | Created       |
| 400  | Bad Request   |
| 404  | Not Found     |
| 500  | Server Error  |


### 💻 Example

```python
import requests

response = requests.get("https://jsonplaceholder.typicode.com/posts/1")

print(response.status_code)  # 200
print(response.json())       # data

## 🔗 REST APIs

### 🧠 Intuition

REST = **standard way to design APIs**

👉 Think:
- URL → resource (what)
- Method → action (what to do)

Example:
- `/users` → users resource
- `/u

### ⚙️ Core Concepts

### Resource-based URLs
- `/users`
- `/users/1`
- `/notes/45`

👉 Always represent **things (nouns)**, not actions


### Stateless

- Each request is independent
- Server does NOT remember previous request

👉 Every request must contain all required data


### JSON Data Format

- Standard data exchange format
- Easy for frontend + backend

Example:
```json
{
  "name": "Dilip",
  "email": "dilip@example.com"
}



### ⚡ FastAPI (Building APIs)


### 🧠 Intuition

FastAPI = **Python framework to build APIs**

👉 Think:
- Express.js (Node.js) but faster
- Built for modern async apps


### ⚙️ Core Concepts

### Route (Endpoint)

- URL where request is sent

Example:
```python
@app.get("/")


### ⚡ Async Programming

### 🧠 Intuition

Async = **non-blocking execution**

👉 Think:
Doing multiple tasks at the same time efficiently

Example:
- Cooking + checking phone + talking


### ⚙️ Core Concepts

### async

- Defines async function

```python
async def task():
