```python
import json
from datetime import datetime


# ---------- 1. GENERATOR (Streaming) ----------
def stream(text):
    for word in text.split():
        yield word


# ---------- 2. MEMORY (Data Layer) ----------
class Memory:
    def __init__(self):
        self.history = []

    def add(self, role, text):
        self.history.append({
            "role": role,
            "text": text,
            "time": datetime.now().strftime("%H:%M:%S")
        })

    def save(self):
        with open("data.json", "w") as f:
            json.dump(self.history, f)

    def load(self):
        try:
            with open("data.json", "r") as f:
                self.history = json.load(f)
        except:
            self.history = []

    def get_context(self):
        return self.history[-3:]  # last 3 messages


# ---------- 3. BRAIN (Logic / Mini GenAI) ----------
class Brain:

    def generate(self, prompt, context):
        prompt = prompt.lower()

        if "hello" in prompt:
            return "Hi! 👋"

        if "time" in prompt:
            return datetime.now().strftime("%H:%M:%S")

        if "genai" in prompt:
            return "GenAI creates text, images, and code."

        # default behavior (uses context)
        context_text = " ".join([msg["text"] for msg in context])
        return f"You said: {prompt}"


# ---------- 4. ASSISTANT (Controller Layer) ----------
class Assistant:

    def __init__(self):
        self.memory = Memory()
        self.brain = Brain()
        self.topics = set()

    def handle(self, user_input):
        # store user input
        self.memory.add("user", user_input)

        # get context
        context = self.memory.get_context()

        # generate response
        response = self.brain.generate(user_input, context)

        # track topic
        if "genai" in user_input.lower():
            self.topics.add("genai")

        # store response
        self.memory.add("bot", response)

        return response


# ---------- 5. MAIN LOOP ----------
def main():
    bot = Assistant()
    bot.memory.load()

    print("Type 'exit' to quit\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            bot.memory.save()
            print("Saved. Bye!")
            break

        response = bot.handle(user_input)

        print("Bot:", end=" ")
        for word in stream(response):
            print(word, end=" ")
        print()


# ---------- 6. ENTRY POINT ----------
if __name__ == "__main__":
    main()
```
