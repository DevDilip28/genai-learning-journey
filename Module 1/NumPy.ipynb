## 📘 NumPy

### 🔹 What is NumPy?

NumPy is a **numerical computing library** in Python used for handling arrays and performing fast mathematical operations.

👉 Think:
- Python list ❌ (slow, limited)
- NumPy array ✅ (fast, optimized)

---

### 🔹 Why NumPy?

Normal Python lists:
- Not efficient for large data
- No built-in vector operations

NumPy:
- Written in C → very fast
- Supports vectorized operations (no loops)
- Used in ML, AI, and data processing

---


```python
import numpy as np

arr = np.array([1, 2, 3])
print(arr)

arr = np.array([1, 2, 3])

print(arr * 2)   # [2, 4, 6]
print(arr + 1)   # [2, 3, 4]
```

```python
np.mean(arr)
np.sum(arr)
np.max(arr)
np.min(arr)
```

```python
import numpy as np

arr = np.array([10, 20, 30])

print("Mean:", np.mean(arr))
print("Multiply:", arr * 5)
```
