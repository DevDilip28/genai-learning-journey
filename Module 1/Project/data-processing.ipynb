```python
import pandas as pd

file_path = input("Enter the file path: ")

df = pd.read_csv(file_path)

print("Select the operation Number you want to perform")
print("1. Display full file content")
print("2. Display specific column content")
print("3. Filter as per condition")

operation = int(input("Enter the operation number: "))

if operation == 1:
    print(df)
elif operation == 2:
    print("Available columns: ", df.columns.tolist())
    column_name = input("Enter the column name: ")
    print(df[column_name])
elif operation == 3:
    condition = input("Enter the condition to filter: ")
    print(df.query(condition))
else:
    print("Invalid operation number")
```
