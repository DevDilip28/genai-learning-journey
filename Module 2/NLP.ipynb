### Stemming - word ko uske root/base form me convert karna (approximate form)

👉 Simple language:

playing → play  
played → play  
plays → play


```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "runs", "easily", "fairly", "happily", "happier", "happiest", "happiness", "happily", "happier", "happiest"]

for word in words:
    print(word, "-->" + stemmer.stem(word))
```

```python
stemmer.stem("better")
```

### Snowball Stemming


```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

words = ["playing", "played", "plays", "studies", "running"]

for word in words:
    print(word, "->", stemmer.stem(word))
```

### Lemmatization - Converts to correct base word

studies → study,

better → good

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "runs", "easily", "fairly", "happily", "happier", "happiest", "happiness"]

for word in words:
    print(word, "-->", lemmatizer.lemmatize(word, pos='v'))
```

### StopWords

Stopwords are common words that are usually removed from text to reduce noise and improve processing efficiency.

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords

tokenizer = TreebankWordTokenizer()

sentence = "I am going to the market and I will buy fruits"

words = tokenizer.tokenize(sentence)

stop_words = set(stopwords.words('english'))

filter_words = [w for w in words if w.lower() not in stop_words]
print(filter_words)
```

```python
corpus = """Artificial Intelligence (AI) is transforming the world rapidly. It is being used in healthcare,
education, finance, and many other industries. AI helps machines to learn from data and make decisions.

Machine Learning is a subset of AI that focuses on building models that can learn patterns from data.
Deep Learning is a further subset that uses neural networks to solve complex problems.

Natural Language Processing (NLP) enables machines to understand human language. It is used in chatbots,
translation systems, and voice assistants like Alexa and Siri.

Data is the new oil in the digital world. The more data we have, the better our models can perform.
However, data must be clean, structured, and relevant.

As technology evolves, ethical concerns around AI are also increasing. It is important to build systems
that are fair, transparent, and accountable.
"""
```

```python
from nltk.tokenize.punkt import PunktSentenceTokenizer
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

sentence_tokenizer = PunktSentenceTokenizer()
sentences = sentence_tokenizer.tokenize(corpus)

word_tokenizer = TreebankWordTokenizer()

all_words = []
for s in sentences:
    all_words.extend(word_tokenizer.tokenize(s))

lemmatizer = WordNetLemmatizer()

base_words = [lemmatizer.lemmatize(word, pos='v') for word in all_words]

vocab = set(all_words)

stop_words = set(stopwords.words('english'))

filtered_words = [w for w in all_words if w.lower() not in stop_words]

print("1 --> ", sentences)
print("2 --> ", all_words)
print("3 --> ", vocab)
print("4 --> ", filtered_words)
print("5 --> ", base_words)
```

### BOW 
Bag of Words (BoW) is a text representation technique used in NLP where a sentence or document is converted into a numerical vector based on the frequency of words, ignoring grammar and word order.

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love AI",
    "I love ML"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("Vectors:\n", X.toarray())
```

### N-Grams
N-gram = sequence of N consecutive words

Types

1️⃣ Unigram (n = 1)

"I love AI" → ["I", "love", "AI"]

2️⃣ Bigram (n = 2)

"I love AI" → ["I love", "love AI"]

3️⃣ Trigram (n = 3)

"I love AI" → ["I love AI"]

```python
from sklearn.feature_extraction.text import CountVectorizer

text = ["I love AI"]

vectorizer = CountVectorizer(ngram_range=(1,2))  # unigram + bigrams
X = vectorizer.fit_transform(text)

print(vectorizer.get_feature_names_out())
```

### 📘 TF-IDF (Term Frequency - Inverse Document Frequency)

TF-IDF is a technique used in NLP to measure how important a word is in a sentence (document) relative to a collection of sentences.



### ⚙️ Formula

### 🔹 Term Frequency (TF)
TF = (Number of times word appears in sentence) / (Total number of words in sentence)

### 🔹 Inverse Document Frequency (IDF)
IDF = log(Total number of sentences / Number of sentences containing the word)

### 🔹 Final Formula
TF-IDF = TF × IDF

---

### 🧪 Example

### Sentences:
- S1: good boy  
- S2: good girl  
- S3: boy girl good  

---

### 🔹 Step 1: Vocabulary

good, boy, girl

---

### 🔹 Step 2: Calculate TF

### For S1 ("good boy")
- TF(good) = 1/2 = 0.5  
- TF(boy) = 1/2 = 0.5  

### For S2 ("good girl")
- TF(good) = 1/2 = 0.5  
- TF(girl) = 1/2 = 0.5  

### For S3 ("boy girl good")
- TF(good) = 1/3  
- TF(boy) = 1/3  
- TF(girl) = 1/3  

---

### 🔹 Step 3: Calculate IDF

Total sentences = 3

- good → appears in 3 sentences  
  IDF(good) = log(3/3) = log(1) = 0  

- boy → appears in 2 sentences  
  IDF(boy) = log(3/2)  

- girl → appears in 2 sentences  
  IDF(girl) = log(3/2)  

---

### 🔹 Step 4: TF-IDF Calculation

### For word "good"
TF-IDF = TF × 0 = 0  

👉 Not important word ❌  

---

### For word "boy"
TF-IDF = TF × log(3/2)  

👉 Important word ✅  

---

### For word "girl"
👉 Same as boy ✅  

---

### 🔥 Key Observations

- Words in all sentences → TF-IDF = 0 ❌  
- Rare words → high TF-IDF ✅  
- Helps identify important words  

---

## 🎯 Summary

- TF → importance in a sentence  
- IDF → importance in dataset  
- TF-IDF → final importance score  

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love AI",
    "I love ML",
    "AI is powerful"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", X.toarray())
```

## Word Embeddings 


## 🔹 Definition
Word Embeddings are numerical vector representations of words such that similar words have similar meanings and similar vectors.

---

## 🧪 Example
"king"  → [0.25, -0.91, 0.73]  
"queen" → [0.27, -0.89, 0.70]  

👉 Similar words → similar vectors  

---

## 🔹 Types

### 1️⃣ Word2Vec
- Learns embeddings using context  
- Two methods:
  - CBOW → context → word  
  - Skip-gram → word → context  

---

### 2️⃣ GloVe
- Uses global word statistics  
- Captures overall word relationships  

---

### 3️⃣ FastText
- Uses subwords (parts of words)  
- Handles unknown words better  

---

### Word2Vec 

Word2Vec is a technique used to create word embeddings by learning from the context of words in sentences.

```python
sentences = [
    ["i", "love", "ai"],
    ["i", "love", "machine", "learning"],
    ["ai", "is", "powerful"]
]
```

```python
from gensim.models import Word2Vec

# training the model
model = Word2Vec(sentences, vector_size=5, window=2, min_count=1)

# get vector of a word
print(model.wv['ai'])

# find similar words 
print(model.wv.most_similar('ai'))
```
