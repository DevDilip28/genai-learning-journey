## 🧠 ML Basics

---

## 🔹 1. What is Machine Learning (ML)

### Explanation
Machine Learning means learning patterns from data instead of writing rules manually.

In traditional programming:
- Rules → Output

In ML:
- Data → Learn patterns → Output

ML models learn relationships between input and output.

---

### Example

Spam Detection:

- "Win money now" → Spam  
- "Meeting at 5 PM" → Not Spam  

Model learns:
- Words like "win", "money" → spam

So for new input:
- "Free money offer" → Spam

👉 Model is not understanding, it is just finding patterns.

---

## 🔹 2. Supervised vs Unsupervised Learning

### Explanation

**Supervised Learning**
- Uses labeled data (input + correct output)
- Learns mapping: Input → Output

**Unsupervised Learning**
- No labels
- Finds patterns or groups in data

---

### Example

**Supervised:**
- "I love this movie" → Positive  
- "Worst experience" → Negative  

Model learns sentiment.

---

**Unsupervised:**
Texts:
- "good movie"
- "great film"
- "bad service"

Model groups:
- Positive texts together  
- Negative texts together  

👉 No labels were given.

---

## 🔹 3. Training vs Inference

### Explanation

**Training**
- Model learns from data
- Happens once or rarely

**Inference**
- Model makes predictions
- Happens when user gives input

---

### Example

Training:
- "good" → positive  
- "bad" → negative  

Model learns pattern.

Inference:
- "great" → predicted positive  

👉 No learning happens here, only prediction.

---

## 🔹 4. Overfitting

### Explanation

Overfitting happens when a model memorizes training data instead of learning general patterns.

Such a model:
- Works well on training data  
- Fails on new data  

---

### Example

Training:
- "I love cats" → Positive  
- "I love dogs" → Positive  

Overfitted model thinks:
- Only these exact sentences are positive

New input:
- "I love animals"

👉 Model fails because it didn’t generalize.

---

# ✅ Final Understanding

- ML = learning from data  
- Supervised = labeled data  
- Unsupervised = no labels  
- Training = learning  
- Inference = predicting  
- Overfitting = memorization (bad)

## 🔹 One Combined Code Example (ML Flow)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Step 1: Data (Supervised Learning)
texts = ["win money", "hello friend", "free offer", "let's meet"]
labels = [1, 0, 1, 0]  # 1 = spam, 0 = not spam

# Step 2: Convert text → numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Step 3: Training (learning phase)
model = MultinomialNB()
model.fit(X, labels)

# Step 4: Inference (prediction phase)
test = vectorizer.transform(["free money"])
prediction = model.predict(test)

print(prediction)  # Output: [1] → Spam
