### Static prompt

```python
from langchain_google_genai import GoogleGenerativeAI

llm = GoogleGenerativeAI(model="gemini-2.5-flash-lite") 

prompt = [
    ("system", "you are a python developer."),
    ("user", "How to sort an array.")
]

res = llm.invoke(prompt)
print(res) 
```

### Dynamic Prompt


```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

llm = GoogleGenerativeAI(model="gemini-2.5-flash-lite")

prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "you are a translator and translate input in {language}."},
    {"role": "user", "content": "{text}"}
])

final_prompt = prompts.format_messages(language="Hindi", text="How are you?")

res = llm.invoke(final_prompt)
print(res) 
```

### Chains


```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate 
from langchain_core.output_parsers import StrOutputParser

llm = GoogleGenerativeAI(model="gemini-2.5-flash-lite")
out = StrOutputParser()

prompt = ChatPromptTemplate.from_messages([
    "Explain {topic} in bullet points in simple one line."
])

def uppercase_parser(result:str):
    return result.upper()

chain = prompt | llm | out | uppercase_parser

res = chain.invoke({"topic": "Explain AI"}) 

print(res) 
```
