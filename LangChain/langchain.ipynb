### Dynamic Prompt


```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

llm = GoogleGenerativeAI(model="gemini-2.5-flash-lite")

prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "you are a translator and translate input in {language}."},
    {"role": "user", "content": "{text}"}
])

final_prompt = prompts.format_messages(language="Hindi", text="How are you?")

res = llm.invoke(final_prompt)
print(res) 
```

```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate 
from langchain_core.output_parsers import StrOutputParser

llm = GoogleGenerativeAI(model="gemini-2.5-flash-lite")
out = StrOutputParser()

prompt = ChatPromptTemplate.from_messages([
    "Explain {topic} in bullet points in simple one line."
])

def uppercase_parser(result:str):
    return result.upper()

chain = prompt | llm | out | uppercase_parser

res = chain.invoke({"topic": "Explain AI"}) 

print(res) 
```

### Memory

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="meta-llama/llama-4-scout-17b-16e-instruct")

history = []

while True:
    query = input("User: ")

    if query.lower() in ["exit", "quit", "bye"]:
        print("Goodbye!")
        break

    history.append({"role": "user", "content": query})

    res = llm.invoke(history)
    
    history.append({"role": "ai", "content": res.content})

    print("AI: ", res.content)
```

```python
# Step 1: Imports
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma


# Step 2: Create Embedding Model
# Converts text → vector (numbers)
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)


# Step 3: Create Sample Data (Knowledge Base)
texts = [
    "LangChain is a framework for building LLM applications",
    "Embeddings convert text into numerical vectors",
    "Vector stores embeddings of data",
    "Retrievers are used to fetch relevant documents"
]


# Step 4: Create Vector Store
# Stores (text, embedding)
vectorstore = Chroma.from_texts(
    texts=texts,
    embedding=embeddings
)


# Step 5: Create Retriever
# Used to search relevant data
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})


# Step 6: User Query
query = "Who is the PM of india?"


# Step 7: Retrieve Relevant Documents
docs = retriever.invoke(query)


# Step 8: Print Results
print("User Query:", query)
print("\nRelevant Results:\n")

for i, doc in enumerate(docs):
    print(f"{i+1}.", doc.page_content)
```

### Retriever fetches relevant data from vector store, and LLM uses that data to generate the final answer.

```python
# Step 1: Imports
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_groq import ChatGroq


# Step 2: Embedding Model
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)


# Step 3: Data (Knowledge Base)
texts = [
    "LangChain is a framework for building LLM applications",
    "Embeddings convert text into numerical vectors",
    "Retrievers fetch relevant documents"
]


# Step 4: Vector Store
vectorstore = Chroma.from_texts(
    texts=texts,
    embedding=embeddings
)


# Step 5: Retriever
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})


# Step 6: LLM (Groq)
llm = ChatGroq(model="meta-llama/llama-4-scout-17b-16e-instruct")


# Step 7: User Query
query = "Who is the Prime Minister of India?"


# Step 8: Retrieve Relevant Docs
docs = retriever.invoke(query)


# Step 9: Combine Context
context = "\n".join([doc.page_content for doc in docs])


# Step 10: Create Final Prompt
final_prompt = f"""
Answer the question based only on the context below:

Context:
{context}

Question:
{query}
"""


# Step 11: LLM Generate Answer
response = llm.invoke(final_prompt)


# Step 12: Output
print("Answer:\n", response.content)
```
