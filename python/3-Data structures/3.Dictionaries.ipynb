### Introduction to Dictionaries

A dictionary is a mutable, insertion-ordered collection of key–value pairs, where each key is unique and immutable, and each value can be of any data type.

```python
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"dilip", "age":21, "grade":"A"}
print(student)
print(type(student))
```

```python
# Single key is always used
student={"name":"Krish","age":32,"name":"dilip"}
print(student)
```

### accessing Dictionary Elements

```python
student={"name":"dilip", "age":21, "grade":"A"}
print(student)
```

```python
## Accessing Dictionary elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

### Modifying Dicitonary Elements

Dictionary are mutable,so you can add, update or delete elements

```python
print(student)
```

```python
student["age"]=23
print(student)

student["location"]="pune"
print(student)
```

```python
del student['grade'] ## delete key and value pair

print(student)
```

```python
## Dictionary methods

keys=student.keys() ##get all the keys
print(keys)

values=student.values() ##get all values
print(values)

items=student.items() ##get all key value pairs
print(items)
```

```python
student={"name":"dilip", "age":21, "grade":"A", "skills": ["Python", "JavaScript"]}
```

```python
## shallow copy

student_copy=student
print(student)
print(student_copy)
```

```python
student["skills"]="c++"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ## shallow copy
print(student_copy1)
print(student)
```

```python
student["skills"]="['Python', 'JavaScript']"
print(student)
print(student_copy1)
```

### Iterating Over Dictionaries

 You can use loops to iterate over dictionatries, keys,values,or items

```python
## Iterate over keys
for keys in student.keys():
    print(keys)

## Iterate over values
for value in student.values():
    print(value)
```

### Nested Disctionaries

```python
students={
    "student1":{"name":"Dilip", "age":21},
    "student2":{"name":"Krish","age":32},
}

print(students)
```

```python
## Access nested dictionaries elementss
print(students["student1"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
for student_id, info in students.items():
    print(f"\n{student_id.upper()}:")
    
    for key, value in info.items():
        print(f"  {key.capitalize()}: {value}")
```

```python
## Dictionary Comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

```python
# Given a sentence, count how many times each word appears — ignoring case and punctuation.

sentence = "Python is great and python is easy, but Python requires practice!"
frequency = {}

# 1️⃣ Convert all to lowercase (so 'Python' and 'python' are same)
sentence = sentence.lower()

# 2️⃣ Remove punctuation (commas, exclamation marks etc.)
for ch in ",.!?":
    sentence = sentence.replace(ch, "")

# 3️⃣ Split sentence into words
words = sentence.split()

# 4️⃣ Count word frequency
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

print(frequency)
```

```python
## USe a dictionary to count he frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4,5]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```
