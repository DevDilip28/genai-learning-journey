### Deep Dive into Operators

Arithmetic Operators

Used to perform basic mathematical operations.

| **Operator** | **Description**        | **Example** | **Output** |
| :----------: | :--------------------- | :---------- | :--------- |
|      `+`     | Addition               | `5 + 3`     | `8`        |
|      `-`     | Subtraction            | `5 - 3`     | `2`        |
|      `*`     | Multiplication         | `5 * 3`     | `15`       |
|      `/`     | Division               | `5 / 2`     | `2.5`      |
|     `//`     | Floor Division         | `5 // 2`    | `2`        |
|      `%`     | Modulus (Remainder)    | `5 % 2`     | `1`        |
|     `**`     | Exponentiation (Power) | `2 ** 3`    | `8`        |


Comparison (Relational) Operators

Used to compare two values.
They return either True or False.

| Operator | Description              | Example  | Output  |
| -------- | ------------------------ | -------- | ------- |
| `==`     | Equal to                 | `5 == 5` | `True`  |
| `!=`     | Not equal to             | `5 != 3` | `True`  |
| `>`      | Greater than             | `5 > 3`  | `True`  |
| `<`      | Less than                | `5 < 3`  | `False` |
| `>=`     | Greater than or equal to | `5 >= 5` | `True`  |
| `<=`     | Less than or equal to    | `3 <= 5` | `True`  |


Logical Operators

Used to combine multiple conditions and return Boolean results.

| Operator | Description                  | Example             | Output  |
| -------- | ---------------------------- | ------------------- | ------- |
| `and`    | True if both are True        | `(x > 5 and y > 5)` | `True`  |
| `or`     | True if at least one is True | `(x > 5 or y < 5)`  | `True`  |
| `not`    | Reverses the result          | `not(x > 5)`        | `False` |


Assignment Operators

Used to assign values or update existing values in variables.

| Operator | Example  | Equivalent To |
| -------- | -------- | ------------- |
| `=`      | `x = 5`  | Assign 5 to x |
| `+=`     | `x += 3` | `x = x + 3`   |
| `-=`     | `x -= 2` | `x = x - 2`   |
| `*=`     | `x *= 3` | `x = x * 3`   |
| `/=`     | `x /= 2` | `x = x / 2`   |
| `%=`     | `x %= 2` | `x = x % 2`   |


String Operators

Used to perform operations on strings.

| Operator | Description   | Example             | Output         |
| -------- | ------------- | ------------------- | -------------- |
| `+`      | Concatenation | `"Hello" + "World"` | `"HelloWorld"` |
| `*`      | Repetition    | `"Hi" * 3`          | `"HiHiHi"`     |


Membership Operators

Used to test whether a value exists in a sequence (like list, tuple, or string).

| Operator | Description               | Example              | Output |
| -------- | ------------------------- | -------------------- | ------ |
| `in`     | True if value is present  | `'a' in 'apple'`     | `True` |
| `not in` | True if value not present | `'b' not in 'apple'` | `True` |


Identity Operators

Used to test whether two variables point to the same object in memory.

| Operator | Description                       | Example      | Output       |
| -------- | --------------------------------- | ------------ | ------------ |
| `is`     | True if both refer to same object | `x is y`     | `True/False` |
| `is not` | True if not same object           | `x is not y` | `True/False` |


```python
## Simple Calculator

# Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Performing arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2
exponentiation = num1 ** num2

# Displaying results
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```
