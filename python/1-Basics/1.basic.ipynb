### Indentation

Indentation in Python is used to define the structure and hierarchy of the code. Unlike many other programming languages that use braces {} to delimit blocks of code, Python uses indentation to determine the grouping of statements. This means that all the statements within a block must be indented at the same level.

```python
age = 18

if age >= 18:
    print("You can vote")
    print("You are an adult")
else:
    print("You are underage")
```

### Type Inference

Type inference means Python automatically detects the data type of a variable based on the value you assign.

In Python, everything is an object — even numbers, strings, lists, and functions.

So, when you write 10, it’s not just a number — it’s an object of the class int.
Similarly, "dilip" is an object of the class str.

```python
variable=10

print(type(variable))

variable="Krish"

print(type(variable))
```
