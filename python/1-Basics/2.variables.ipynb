### Variables

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
# Declaring and Assigning Variables

name = "Dilip"
age = 21
height = 6.0
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)
```

## üß† Naming Conventions 

Variable names should be descriptive.

Must start with a letter or an underscore (_).

Can contain letters, numbers, and underscores.

Variable names are case-sensitive (name ‚â† Name).

#### ‚úÖ Valid variable names
first_name = "Krish"

last_name = "Naik"

_age = 25

user2 = "Dilip"

#### ‚ùå Invalid variable names
2age = 30       # Cannot start with a number

first-name = "Krish"  # Hyphens are not allowed

@name = "Krish"       # Special characters are not allowed

### üß† Understanding Variable Types
- Python is **dynamically typed**, meaning variable types are decided at runtime.  
- Example:
  age = 25        # int  
  height = 6.1    # float  
  name = "Krish"  # str  
  is_student = True # bool  


```python
age = 25
print(type(age))         

age_str = str(age)
print(age_str)          
print(type(age_str))     

age = '25'
print(type(int(age)))   

name = "Krish"
int(name)  # ‚ùå Error ‚Äì cannot convert non-numeric string
```
