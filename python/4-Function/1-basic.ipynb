Definition:

A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.


```python
## syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
def even_or_odd(num):
    """Function to check number is even or odd"""

    if(num%2 == 0):
        return "even"
    else:
        return "odd"

int = int(input("Enter the number: "))
result = even_or_odd(int)
print(result)
```

```python
def add(a, b):
    return a+b

sum = add(5, 10)
print(sum)
```

```python
## Default Parameters

def greet(name="guest"):
    print(f"Hello {name}, welcome to the world of Python!")

greet("Dilip")
```

```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_number(*args):
    for i in args:
        print (i)

print_number(1, 2, 3, 4, 5, "Dilip")
```

```python
### Keywords Arguments

def details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

details(name="Dilip", age=21, city="Pune")
```

```python
def fun(a=10, *args, **kwargs):
    print(a, args, kwargs)

fun(5, 1, 2, name="Dilip")
```
