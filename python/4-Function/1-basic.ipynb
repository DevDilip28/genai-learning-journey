A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

```python
def even_or_odd(num):
    """Function to check number is even or odd"""

    if(num%2 == 0):
        return "even"
    else:
        return "odd"

int = int(input("Enter the number: "))
result = even_or_odd(int)
print(result)
```

```python
def add(a, b):
    return a+b

sum = add(5, 10)
print(sum)
```
