Example 1: Temperature Conversion

```python
def temp_conversion(temp, unit):
    if unit == "c":
        return temp * 9/5 + 32
    elif unit == "f":
        return temp - 32 * 5/9
    else:
        return None
    
print(temp_conversion(25,'c'))
print(temp_conversion(77,'f'))
```

Example 2: Password Strength Checker

```python
def is_strong_password(password: str) -> bool:
    special_chars = set("!@#$%^&*()_+")

    return (
        len(password) >= 8
        and any(c.isdigit() for c in password)
        and any(c.islower() for c in password)
        and any(c.isupper() for c in password)
        and any(c in special_chars for c in password)
    )


print(is_strong_password("WeakPwd"))      
print(is_strong_password("Str0ngPwd!"))   
```

Example 3: Calculate the Total Cost


```python
def calculate_toral_price(cart):
    total = 0
    for item in cart:
        total = total + item['price']*item['quantity']
    return total

cart = [
    {'name': 'apple', 'price': 0.5, 'quantity': 4},
    {'name': 'banana', 'price': 0.3, 'quantity': 6},
    {'name': 'orange', 'price': 0.8, 'quantity': 3}
]

calculate_toral_price(cart)
```

Example 4: Check IF a String Is Palindrome

```python
def is_palindrome(s: str) -> bool:
    s = s.lower().replace(" ", "")
    
    return s == s[::-1]

print(is_palindrome("heelo"))
print(is_palindrome("A man a plan a canal Panama")) 
```
