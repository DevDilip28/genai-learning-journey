## The map() Function
The map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
    return x*x

square(10)
```

```python
number = [1,2,3,4,5]

list(map(square, number))
```

```python
list(map(lambda x:x*x, number))
```

```python
row1 = [1,2,3,4,5]
row2 = [6,7,8,9,10]

add_numbers = list(map(lambda x,y : x+y, row1, row2))

print(add_numbers)
```

```python
str_numbers = ['1','2','3','4','5']

list(map(int, str_numbers))
```

```python
fruits = ['apple', 'banana', 'cherry', 'date']

list(map(lambda x: x.upper(), fruits))
```

```python
def get_name(person):
    return person['name']

people = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35}
    ]

list(map (get_name, people))
```
