```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_groq import ChatGroq
from langchain_chroma import Chroma
from langchain_core.prompts import PromptTemplate

loaders = PyPDFLoader("./Data/LLM.pdf")
docs = loaders.load()

splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splitted_data = splitter.split_documents(docs) 

embedding = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en")

vectore_store = Chroma.from_documents(
    documents=splitted_data,
    embedding=embedding
)

```

```python
query = "What are types of AI"

data = vectore_store.similarity_search(query=query)

context = ""

for doc in data:
    context += doc.page_content + "\n"


llm = ChatGroq(model="meta-llama/llama-4-scout-17b-16e-instruct")

# res = llm.invoke(f"Provide answer based on context, Context: {context}, and question is {query}")
# print(res.content) 
```

### chain - context generate | prompt | llm | strparser

```python
def get_context(query:str):
    data = vectore_store.similarity_search(query=query)

    context = ""

    for doc in data:
        context += doc.page_content + "\n"

    return{
        "context": context,
        "question": query
    }

prompt = PromptTemplate.from_template("""
    You are a helpfull assistant and answer based on the context and if you dont know the answer then you can say 'I don't know'.
    Context: {context}
    Question: {question}
""")
```

```python
rag_chain = get_context | prompt | llm 

res = rag_chain.invoke("What is LLM")

print(res.content)
```
