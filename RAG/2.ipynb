### Chunking

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = "Artificial Intelligence (AI) is the simulation of human intelligence processes by machines, especially computer systems. These processes include learning, reasoning, and self-correction."

splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=20)

chunks = splitter.split_text(text)

chunks
```

```python
from langchain_community.document_loaders import TextLoader, PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter 

text_loader = TextLoader("./Data/AI.txt")
pdf_loader = PyPDFLoader("./Data/LLM.pdf")

text = text_loader.load()
pdf = pdf_loader.load()

splitter = RecursiveCharacterTextSplitter(chunk_size=60, chunk_overlap=20)

text_chunks = splitter.split_documents(text)
pdf_chunks = splitter.split_documents(pdf)

print(f"Text {len(text_chunks)}")
print(f"PDF {len(pdf_chunks)}") 
```
